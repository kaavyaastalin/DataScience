```python
#importing the headers
import numpy as np
import matplotlib.pyplot as mat
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation#loading the dataset
```

```python
#loading the dataset and converting to array format
dataset = pd.read_csv('Desktop\labeled_data.csv',header=None).values
dataset
```

```python
#splitting the dataset into  X (independant variables) and y (dependant variable)
X = dataset[:,:-1] #all columns except the last one
y = dataset[:,len(dataset[0])-1] #only the last column
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()
```

```python
# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)
```

```python
#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO  
from IPython.display import Image
import pydotplus

dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('datascience1.png')
Image(graph.create_png())
```

```python
#selecting 1st 15 columns,lables columns and name it as feature(1..15),ID to apply other kind of split algorithm in Desicion Tree
df = pd.read_csv("Desktop\ROndom_labeled_data.csv")
df.head()
```

```python
#split dataset in features and target variable
feature_cols = ['Feature1', 'Feature2', 'Feature3', 'Feature4','Feature5','Feature6','Feature7','Feature8', 'Feature9', 'Feature10', 'Feature11','Feature12','Feature13','Feature14','Feature15']
X = df[feature_cols] # Features
y = df.ID # Target variable
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()
```

```python
# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)
```

```python
#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
# Model Accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
